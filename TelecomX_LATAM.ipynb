# Extracci贸n

```python
import pandas as pd
import requests

print("--- Fase 1: Extracci贸n de Datos desde la API ---")

url = "https://github.com/alura-cursos/challenge2-data-science-LATAM/raw/main/TelecomX_Data.json"

response = requests.get(url)

if response.status_code == 200:
    print("Conexi贸n exitosa. C贸digo de estado:", response.status_code)
    datos_brutos = pd.json_normalize(data=response.json())

    print(f"隆Extracci贸n completada! Se han cargado {len(datos_brutos)} registros de clientes.")
    print("\nMostrando las primeras 5 filas de los datos extra铆dos:")
    display(datos_brutos.head())

else:
    print("Error al conectar con la API. C贸digo de estado:", response.status_code)
```

# Transformaci贸n

```python
print(" Informaci贸n General del DataFrame ")
datos_brutos.info()

print("\ Dimensiones del DataFrame (Filas, Columnas) ")
print(datos_brutos.shape)
```

```python
# Lista de columnas categ贸ricas que se van a revisar
columnas_a_revisar = ['Churn', 'customer.gender', 'customer.Partner', 'customer.Dependents', 'account.PaymentMethod']

print("Revisando Valores nicos en Columnas Categ贸ricas ")

for columna in columnas_a_revisar:
  valores_unicos = datos_brutos[columna].unique()
  print(f"\nValores 煤nicos en la columna '{columna}':")
  print(valores_unicos)

# Verificaion de nuestra columna sospechosa
print("\nInvestigando la columna 'account.Charges.Total'")
print("Valores 煤nicos iniciales (primeros 10):")
print(datos_brutos['account.Charges.Total'].unique()[:10])
```

```python
# errors='coerce' transformar谩 cualquier valor no num茅rico en 'NaN' (Not a Number).
cargos_totales_numericos = pd.to_numeric(datos_brutos['account.Charges.Total'], errors='coerce')

# Ahora, buscamos las filas donde la conversi贸n fall贸 (donde el resultado es NaN).
filas_con_errores = datos_brutos[cargos_totales_numericos.isnull()]

# Se mostrara las filas problem谩ticas para ver qu茅 contienen
print(f"Se encontraron {len(filas_con_errores)} filas con valores no num茅ricos en 'account.Charges.Total'.")
print("\nMostrando las filas con inconsistencias:")
display(filas_con_errores)
```

```python
# Contamos cu谩ntos valores tiene problema (' ') antes de la limpieza
conteo_antes = (datos_brutos['account.Charges.Total'] == ' ').sum()
print(f"Valores problem谩ticos (' ') antes de la limpieza: {conteo_antes}")

# Se reemplazaran los espacios en blanco con '0' .
datos_brutos.loc[datos_brutos['account.Charges.Total'] == ' ', 'account.Charges.Total'] = '0'

# Se vrifica el conteo
conteo_despues = (datos_brutos['account.Charges.Total'] == ' ').sum()
print(f"Valores problem谩ticos (' ') despu茅s de la limpieza: {conteo_despues}")

# Ahora que no hay texto, convertimos la columna a tipo num茅rico (float)
datos_brutos['account.Charges.Total'] = datos_brutos['account.Charges.Total'].astype(float)

# Verificaci贸n final: mostramos la informaci贸n del DataFrame de nuevoo'
print("\n--- Verificaci贸n del DataFrame despu茅s de la limpieza ---")
datos_brutos.info()
```

```python
# Definimos el n煤mero de d铆as en un mes para el c谩lculo
DIAS_MES = 30

# 1. Creamos la nueva columna 'Cuentas_Diarias'; Se calcula dividiendo los cargos mensuales por 30
datos_brutos['Cuentas_Diarias'] = datos_brutos['account.Charges.Monthly'] / DIAS_MES

# 2. Verificamos que la columna fue creada correctamente y se usara .head() para ver las primeras filas y nuestra nueva columna al final
print("--- Verificaci贸n de la nueva columna 'Cuentas_Diarias' ---")
print("Mostrando las primeras 5 filas con la columna a帽adida:")
display(datos_brutos.head())

# Se verificar con .info() para ver que ahora tenemos 22 columnas
print("\n--- Informaci贸n actualizada del DataFrame ---")
datos_brutos.info()
```

```python
datos_transformados = datos_brutos.copy()

mapa_si_no = {'Yes': 1, 'No': 0}

# Columnas que tienen valores 'Yes'/'No'
columnas_si_no = [
    'Churn',
    'customer.Partner',
    'customer.Dependents',
    'phone.PhoneService',
    'phone.MultipleLines',
    'internet.OnlineSecurity',
    'internet.OnlineBackup',
    'internet.DeviceProtection',
    'internet.TechSupport',
    'internet.StreamingTV',
    'internet.StreamingMovies',
    'account.PaperlessBilling'
]

# Usamos un bucle for para aplicar la transformaci贸n a todas las columnas de la lista
for columna in columnas_si_no:
    datos_transformados[columna] = datos_transformados[columna].map(mapa_si_no)

# Hacemos lo mismo para la columna de g茅nero por separado, ya que tiene otros valores
datos_transformados['customer.gender'] = datos_transformados['customer.gender'].map({'Male': 1, 'Female': 0})

# Verificamos los cambios viendo las primeras filas
print("--- Verificaci贸n de Datos Estandarizados (Convertidos a 1s y 0s) ---")
display(datos_transformados.head())

# Verificamos los nuevos tipos de datos
print("\n--- Informaci贸n actualizada del DataFrame Transformado ---")
datos_transformados.info()
```
