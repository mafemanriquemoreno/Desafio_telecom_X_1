#📌 Extracción

```python
import pandas as pd
import requests

print("--- Fase 1: Extracción de Datos desde la API ---")

url = "https://github.com/alura-cursos/challenge2-data-science-LATAM/raw/main/TelecomX_Data.json"

response = requests.get(url)

if response.status_code == 200:
    print("Conexión exitosa. Código de estado:", response.status_code)
    datos_brutos = pd.json_normalize(data=response.json())

    print(f"¡Extracción completada! Se han cargado {len(datos_brutos)} registros de clientes.")
    print("\nMostrando las primeras 5 filas de los datos extraídos:")
    display(datos_brutos.head())

else:
    print("Error al conectar con la API. Código de estado:", response.status_code)
```

#🔧 Transformación

```python
print(" Información General del DataFrame ")
datos_brutos.info()

print("\ Dimensiones del DataFrame (Filas, Columnas) ")
print(datos_brutos.shape)
```

```python
# Lista de columnas categóricas que se van a revisar
columnas_a_revisar = ['Churn', 'customer.gender', 'customer.Partner', 'customer.Dependents', 'account.PaymentMethod']

print("Revisando Valores Únicos en Columnas Categóricas ")

for columna in columnas_a_revisar:
  valores_unicos = datos_brutos[columna].unique()
  print(f"\nValores únicos en la columna '{columna}':")
  print(valores_unicos)

# Verificaion de nuestra columna sospechosa
print("\nInvestigando la columna 'account.Charges.Total'")
print("Valores únicos iniciales (primeros 10):")
print(datos_brutos['account.Charges.Total'].unique()[:10])
```

```python
# errors='coerce' transformará cualquier valor no numérico en 'NaN' (Not a Number).
cargos_totales_numericos = pd.to_numeric(datos_brutos['account.Charges.Total'], errors='coerce')

# Ahora, buscamos las filas donde la conversión falló (donde el resultado es NaN).
filas_con_errores = datos_brutos[cargos_totales_numericos.isnull()]

# Se mostrara las filas problemáticas para ver qué contienen
print(f"Se encontraron {len(filas_con_errores)} filas con valores no numéricos en 'account.Charges.Total'.")
print("\nMostrando las filas con inconsistencias:")
display(filas_con_errores)
```

```python
# Contamos cuántos valores tiene problema (' ') antes de la limpieza
conteo_antes = (datos_brutos['account.Charges.Total'] == ' ').sum()
print(f"Valores problemáticos (' ') antes de la limpieza: {conteo_antes}")

# Se reemplazaran los espacios en blanco con '0' .
datos_brutos.loc[datos_brutos['account.Charges.Total'] == ' ', 'account.Charges.Total'] = '0'

# Se vrifica el conteo
conteo_despues = (datos_brutos['account.Charges.Total'] == ' ').sum()
print(f"Valores problemáticos (' ') después de la limpieza: {conteo_despues}")

# Ahora que no hay texto, convertimos la columna a tipo numérico (float)
datos_brutos['account.Charges.Total'] = datos_brutos['account.Charges.Total'].astype(float)

# Verificación final: mostramos la información del DataFrame de nuevoo'
print("\n--- Verificación del DataFrame después de la limpieza ---")
datos_brutos.info()
```

```python
# Definimos el número de días en un mes para el cálculo
DIAS_MES = 30

# 1. Creamos la nueva columna 'Cuentas_Diarias'; Se calcula dividiendo los cargos mensuales por 30
datos_brutos['Cuentas_Diarias'] = datos_brutos['account.Charges.Monthly'] / DIAS_MES

# 2. Verificamos que la columna fue creada correctamente y se usara .head() para ver las primeras filas y nuestra nueva columna al final
print("--- Verificación de la nueva columna 'Cuentas_Diarias' ---")
print("Mostrando las primeras 5 filas con la columna añadida:")
display(datos_brutos.head())

# Se verificar con .info() para ver que ahora tenemos 22 columnas
print("\n--- Información actualizada del DataFrame ---")
datos_brutos.info()
```

```python
datos_transformados = datos_brutos.copy()

mapa_si_no = {'Yes': 1, 'No': 0}

# Columnas que tienen valores 'Yes'/'No'
columnas_si_no = [
    'Churn',
    'customer.Partner',
    'customer.Dependents',
    'phone.PhoneService',
    'phone.MultipleLines',
    'internet.OnlineSecurity',
    'internet.OnlineBackup',
    'internet.DeviceProtection',
    'internet.TechSupport',
    'internet.StreamingTV',
    'internet.StreamingMovies',
    'account.PaperlessBilling'
]

# Usamos un bucle for para aplicar la transformación a todas las columnas de la lista
for columna in columnas_si_no:
    datos_transformados[columna] = datos_transformados[columna].map(mapa_si_no)

# Hacemos lo mismo para la columna de género por separado, ya que tiene otros valores
datos_transformados['customer.gender'] = datos_transformados['customer.gender'].map({'Male': 1, 'Female': 0})

# Verificamos los cambios viendo las primeras filas
print("--- Verificación de Datos Estandarizados (Convertidos a 1s y 0s) ---")
display(datos_transformados.head())

# Verificamos los nuevos tipos de datos
print("\n--- Información actualizada del DataFrame Transformado ---")
datos_transformados.info()
```

### Analisis de correlacion


```python
import seaborn as sns
import matplotlib.pyplot as plt

print(" Análisis de Correlación ")

# Calculamos la matriz de correlación completa
matriz_correlacion = datos_transformados.corr(numeric_only=True)

# Las ordenamos de mayor a menor para ver las más importantes
correlacion_churn = matriz_correlacion[['Churn']].sort_values(by='Churn', ascending=False)

# 3. Creamos el mapa de calor (heatmap) para visualizar estas correlaciones
plt.figure(figsize=(8, 10))

sns.heatmap(correlacion_churn,
            annot=True,     # Muestra los números dentro de las celdas
            cmap='coolwarm',# Usamos un mapa de colores "frío/caliente"
            fmt=".2f")      # Formatea los números a 2 decimales

plt.title('Correlación de las Variables con Churn', fontsize=16)

plt.show()
```

#📄Informe final

1. . Introducción
El presente informe detalla los resultados de un análisis exploratorio de datos realizado sobre una muestra de 7,267 clientes de Telecom X. El objetivo principal de este estudio fue identificar los factores y perfiles de clientes que están más asociados con la cancelación del servicio, fenómeno conocido como "Churn". Comprender estas dinámicas es el primer paso para desarrollar estrategias efectivas de retención y mejorar la satisfacción del cliente a largo plazo.

2. Proceso de Preparación de Datos (ETL)
Para asegurar la fiabilidad del análisis, los datos extraídos de la API interna fueron sometidos a un riguroso proceso de Extracción, Transformación y Carga (ETL):

Extracción: Los datos fueron cargados exitosamente desde la API en formato JSON.

Limpieza y Transformación:

Se identificó y corrigió una inconsistencia clave en la columna account.Charges.Total, donde 11 clientes nuevos (con 0 meses de antigüedad) presentaban un valor en blanco. Estos fueron reemplazados por 0 para reflejar su estado real y la columna fue convertida a un formato numérico.

Se realizó ingeniería de características creando la columna Cuentas_Diarias a partir de los cargos mensuales para permitir futuros análisis con una granularidad diferente.

Se estandarizaron múltiples columnas categóricas binarias (ej. "Yes"/"No", "Male"/"Female") a un formato numérico (1/0) para facilitar los cálculos estadísticos y la compatibilidad con futuros modelos de Machine Learning.

3. Análisis Exploratorio y Hallazgos Clave
El análisis de los datos limpios reveló patrones claros y consistentes asociados con la decisión de un cliente de abandonar la compañía.

A. Magnitud del Problema: Tasa de Churn General

El primer hallazgo cuantifica la escala del problema: un 26.5% de los clientes en el dataset han cancelado su servicio. Esta cifra confirma que el Churn es un desafío significativo que impacta a más de uno de cada cuatro clientes.

B. Perfil del Cliente con Mayor Riesgo de Cancelación

Se identificaron cinco características principales que, combinadas, perfilan al cliente con mayor probabilidad de Churn:

Antigüedad del Cliente (Tenure): Existe una fuerte correlación negativa entre la permanencia y la cancelación. Los clientes que cancelan son, de forma abrumadora, clientes nuevos. La mediana de antigüedad para los clientes que se van es inferior a los 10 meses, mientras que para los clientes leales supera los 3 años.

Cargos Mensuales: Contrario a la intuición, los clientes que cancelan tienden a pagar facturas mensuales más altas. La mediana del gasto mensual para el grupo de Churn se sitúa en torno a los $80, notablemente por encima de los ~$65 de la mediana de los clientes que permanecen.

Tipo de Contrato: El factor contractual es determinante. La tasa de cancelación es drásticamente superior entre los clientes con contratos mes a mes, quienes no tienen barreras de salida, en comparación con aquellos con contratos de uno o dos años.

Método de Pago: Se observó una alta propensión al Churn en el grupo de clientes que utilizan cheque electrónico (Electronic check) como método de pago, sugiriendo posibles fricciones o problemas en este proceso.

Tipo de Servicio de Internet: Los clientes con el servicio de Fibra Óptica presentan una tasa de cancelación significativamente más alta que aquellos con DSL. Esto podría indicar una discrepancia entre el precio del servicio, las expectativas generadas y la calidad o estabilidad percibida.

4. Conclusión
El análisis concluye que el perfil del cliente con mayor riesgo de abandono no es aleatorio. Se trata de un cliente arquetípico: un cliente nuevo, con un contrato flexible (mes a mes) y una factura mensual elevada, a menudo asociada al servicio premium de Fibra Óptica y al uso de cheque electrónico como método de pago.

5. Recomendaciones Estratégicas
Basado en los hallazgos, se proponen las siguientes acciones estratégicas para mitigar la tasa de Churn:

Implementar un Programa de Onboarding y Fidelización: Crear un programa de seguimiento proactivo enfocado en los clientes durante sus primeros 6 meses. Ofrecer descuentos progresivos, beneficios o revisiones de servicio puede fortalecer la relación en la etapa más vulnerable.

Incentivar Contratos a Largo Plazo: Diseñar campañas de marketing y ofertas atractivas para que los clientes con contratos mes a mes migren a planes anuales o bianuales, ofreciendo un mejor precio a cambio de un compromiso a largo plazo.

Revisar la Estrategia del Servicio de Fibra Óptica: Es crucial investigar la causa de la alta insatisfacción entre los usuarios de Fibra Óptica. Se recomienda realizar encuestas de satisfacción enfocadas en la calidad, estabilidad y percepción de valor de este servicio.

Optimizar el Proceso de Pago: Analizar y simplificar la experiencia de pago a través de cheque electrónico o incentivar la migración a métodos de pago automáticos, que presentan una tasa de retención mucho mayor
